# Basic approach for fizz-buzz

```python
for Codingeeks in range(1,100): 
    if Codingeeks % 3 == 0 and Codingeeks % 5 == 0: 
        print("Codingeeks")                                         
        continue
 
    elif Codingeeks % 3 == 0:     
        print("Coding")                                         
        continue

    elif Codingeeks % 5 == 0:         
        print("geeks")                                     
        continue

    print(Codingeeks)
```

# Average approach for fizz-buzz

```python
for Codingeeks in range(1,100): 
    if Codingeeks % 15 == 0 : 
        print("Codingeeks")                                         
        continue
 
    elif Codingeeks % 3 == 0:     
        print("Coding")                                         
        continue

    elif Codingeeks % 5 == 0:         
        print("geeks")                                     
        continue

    print(Codingeeks)
```

# Advanced approach for fizz-buzz

```python
for Codingeeks in range(1,100):
    CG=""
    if Codingeeks%3 == 0: 
        CG = CG + 'Coding'
    if Codingeeks%5 == 0:
        CG = CG + 'geeks'
    if CG=="":
        print(Codingeeks)
    else:
        print(CG)
        
```

# Pro approach for fizz-buzz

```python
Coding = 0
geeks = 0
for Codingeeks in range(1,100):

    Coding = Coding+1
    geeks = geeks+1     
    CG =""
    if (Coding ==3):
        CG = CG+'Coding'
        Coding = 0
    if ( geeks==5 ):
        CG = CG +'geeks'
        geeks = 0
    if CG == "":
        print(Codingeeks)
    else:
        print(CG)
```
